```python
import numpy as np
import matplotlib.pyplot as plt
```

# Model

```python
def Model(a, b, c, d, dt, t):
    timestamp = np.arange(0, t * (1 / dt), dt)
    N = timestamp.size

    U = np.zeros(N)
    v = np.zeros(N)

    v[0] = c
    U[0] = c * b

    I = 10.0 * np.ones(N)

    for t in range(N - 1):
        if v[t] >= 30:
            v[t] = c
            U[t] = U[t] + d

        v[t + 1] = (0.04 * pow(v[t], 2) + 5.0 * v[t] + 140.0 - U[t] + I[t]) * dt + v[t]
        U[t + 1] = (a * (b * v[t] - U[t])) * dt + U[t]

    return v, timestamp
    
    
```

# Intrinsically Bursting (IB)

```python
V, T = Model(0.02, 0.2, -55.0, 4.0, 0.1, 10)
plt.plot(T, V)
plt.show()
```

# Chattering (CH)

```python
V, T = Model(0.02, 0.2, -50.0, 2.0, 0.1, 10)
plt.plot(T, V)
plt.show()
```

# Fast Spiking (FS)
