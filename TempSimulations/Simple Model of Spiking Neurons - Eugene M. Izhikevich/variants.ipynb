```python
import numpy as np
import matplotlib.pyplot as plt
```

# Model

```python
def Model(a, b, c, d, dt, t):
    T = np.arange(0, t * (1 / dt), dt)
    N = T.size

    U = np.zeros(N)
    V = np.zeros(N)

    V[0] = c
    U[0] = c * b

    I = 10.0 * np.ones(N)

    for t in range(N - 1):
        if V[t] >= 30:
            V[t] = c
            U[t] = U[t] + d

        V[t + 1] = (0.04 * pow(V[t], 2) + 5.0 * V[t] + 140.0 - U[t] + I[t]) * dt + V[t]
        U[t + 1] = (a * (b * V[t] - U[t])) * dt + U[t]

    return V, T
    
    
```

# Regular Spiking (RS)

```python
V, T = Model(0.02, 0.2, -65.0, 8.0, 0.1, 50)
plt.plot(T, V)
plt.show()
```

# Intrinsically Bursting (IB)

```python
V, T = Model(0.02, 0.2, -55.0, 4.0, 0.1, 10)
plt.plot(T, V)
plt.show()
```

# Chattering (CH)

```python
V, T = Model(0.02, 0.2, -50.0, 2.0, 0.1, 10)
plt.plot(T, V)
plt.show()
```

# Fast Spiking (FS)

```python
V, T = Model(0.03, 0.2, -65.0, 2.0, 0.1, 10)
plt.plot(T, V)
plt.show()
```

# Thalamo-cortical (TC)

```python
V, T = Model(0.02, 0.25, -65.0, 0.05, 0.1, 2)
plt.plot(T, V)
plt.show()
```
